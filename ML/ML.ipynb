```python
import pandas as pd
import numpy as np
import tensorflow as tf

data = pd.read_csv("output.csv", header=0, delimiter=", ", decimal=".", engine='python')
```

```python
local = pd.read_csv("AE_local_no_anomalies.csv", header=0, delimiter=",", decimal=".", engine='python')
cloud = pd.read_csv("AE_cloud_no_anomalies.csv", header=0, delimiter=",", decimal=".", engine='python')
cloud.head()
```

```python
cloud = data[data['type'] == "cloud"]
local = data[data['type'] == "local"]
```

```python
cloud['battery'] = cloud['battery'].apply(lambda x: int(x) * -1)
cloud['time'] = cloud['time'].div(data['iterations'], axis = 0)

cloud_time = pd.to_numeric(cloud.time)
cloud_size = pd.to_numeric(cloud.size2)
cloud_battery = pd.to_numeric(cloud.battery)
cloud_pages = pd.to_numeric(cloud.pages)
```

```python
X_cloud = pd.DataFrame({'size': cloud_size, 'pages': cloud_pages}) 
Y_cloud = pd.DataFrame({'battery': cloud_battery, 'time': cloud_time}) 
```

```python
local['battery'] = local['battery'].apply(lambda x: int(x) * -1)
local['time'] = local['time'].div(data['iterations'], axis = 0)

local_time = pd.to_numeric(local.time)
local_size = pd.to_numeric(local.size2)
local_battery = pd.to_numeric(local.battery)
local_pages = pd.to_numeric(local.pages)
```

```python
X_local = pd.DataFrame({'size': local_size, 'pages': local_pages}) 
Y_local = pd.DataFrame({'battery': local_battery, 'time': local_time}) 
```

```python
from tensorflow import keras
from tensorflow.keras import layers
model_cloud = keras.Sequential()
model_cloud.add(layers.Dense(50, input_dim=2, kernel_initializer='he_uniform', activation='relu'))
model_cloud.add(layers.Dense(2))
model_cloud.compile(loss='mae', optimizer='adam', metrics=[keras.metrics.MeanAbsoluteError()])

model_local = keras.Sequential()
model_local.add(layers.Dense(50, input_dim=2, kernel_initializer='he_uniform', activation='relu'))
model_local.add(layers.Dense(2))
model_local.compile(loss='mae', optimizer='adam', metrics=[keras.metrics.MeanAbsoluteError()])
```

```python
from sklearn.model_selection import train_test_split
X_train_local, X_test_local, y_train_local, y_test_local = train_test_split(X_local, Y_local, test_size=0.20, random_state=42)
X_train_cloud, X_test_cloud, y_train_cloud, y_test_cloud = train_test_split(X_cloud, Y_cloud, test_size=0.20, random_state=42)
```

```python
history_cloud = model_cloud.fit(X_train_cloud, y_train_cloud, batch_size=8, epochs=300, verbose=1, validation_split=0.2)
history_local = model_local.fit(X_train_local, y_train_local, batch_size=8, epochs=300, verbose=1, validation_split=0.2)
```

```python
score_cloud, accuracy_cloud = model_cloud.evaluate(X_test_cloud, y_test_cloud, verbose=1)

print("Test Score:", score_cloud)
print("Accuracy:", accuracy_cloud)
```

```python
score_local, accuracy_local = model_local.evaluate(X_test_local, y_test_local, verbose=1)

print("Test Score:", score_local)
print("Accuracy:", accuracy_local)
```

```python
converter_cloud = tf.lite.TFLiteConverter.from_keras_model(model_cloud)
converter_cloud.inference_output_type = tf.int8
converter_cloud.optimizations = [tf.lite.Optimize.DEFAULT]
tflite_model_cloud = converter_cloud.convert()
open("AE_cloud.tflite", "wb").write(tflite_model_cloud)

converter_local = tf.lite.TFLiteConverter.from_keras_model(model_local)
converter_local.inference_output_type = tf.int8
converter_local.optimizations = [tf.lite.Optimize.DEFAULT]
tflite_model_local = converter_local.convert()
open("AE_local.tflite", "wb").write(tflite_model_local)
```
